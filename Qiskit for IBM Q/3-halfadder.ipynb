```python
from qiskit import *
```

```python
qr = QuantumRegister(4)

cr = ClassicalRegister(2)
```

```python
qc = QuantumCircuit(qr,cr)

qc.draw(output='mpl')
```

```python
qc.x([0,1]) #NOT gate
qc.draw(output='mpl')
```

```python
qc.cx(0,2) # Controlled NOT gate
qc.draw(output='mpl')
```

```python
qc.cx(1,2)
qc.draw(output='mpl')
```

```python
qc.ccx(0,1,3) # Controlled-controlled NOT gate TOFFOLI GATE
qc.draw(output='mpl')
```

```python
qc.measure(2,0)
qc.measure(3,1)
qc.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, backend=simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram 
```

```python
plot_histogram(result.get_counts(qc))
```
